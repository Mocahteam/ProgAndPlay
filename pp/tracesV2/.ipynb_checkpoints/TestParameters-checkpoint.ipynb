```python
import subprocess
import os 
import matplotlib.pyplot as plt
import numpy as np
import operator
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
def read_solution(file):
    ret = []
    with open(file) as f:
        while True:
            lines = f.readline()
            if not lines:
                break
            ret.append(lines[0:-1])
    return ret


def check_solution_correct(file,solution):
    num=0
    last_score = "***"
    with open(file) as f:
        while True:
            lines = f.readline()
            if not lines:
                break
            if (lines[0]!="["):
                continue
            num += 1
            line = lines.split(" ")
            best = []
            for l in line[8:]:
                if("execution_end_time" in l or "mission_end_time" in l or "EndRoot" in l):
                    break
                best.append(l)
            if(line[1]==last_score):
                num -= 1
            last_score = line[1]
            #print("solution ",solution,"\n")
            #print("best ",best,"\n")
            if (operator.eq(best,solution)):
                #print("Really!")
                return num;
            
        return 2000;
    
def search_score(debug_file, solution):
    try:
        with open(debug_file) as f:
            while True:
                lines = f.readline()
                if not lines:
                    break
                if (lines[0]!="["):
                    continue
                line = lines.split(" ")
                best = []
                for l in line[8:]:
                    if("execution_end_time" in l or "mission_end_time" in l or "EndRoot" in l):
                        break
                    best.append(l)
                #print("Trying : Best=",best, " Solution=",solution)
                if (operator.eq(best,solution)):
                    return float(line[1])
            return 0
    
    except FileNotFoundError:
        print("File not found! -> ",debug_file)
        
def dessin3D(data, file="#", dynamic=False, war_from=0, war_to=1, wma_from=0, wma_to=1, wml_from=0, wml_to=1, \
             only_red=False, only_black=False, without_black=False,\
             only_yellow=False, sommeA1=False, use_score=True, debug_folder="./files_debug", tf=""):
    if(dynamic):
        %matplotlib notebook
    else:
        %matplotlib inline
    J = np.array([[1,1,1],[0,1,0.5],[0,0,(3**0.5)/2]])
        
    fig = plt.figure()
    ax1 = plt.axes(projection='3d')
    war_l = []
    wma_l = []
    wml_l = []
    couleur_l = []
    count = 0
    if use_score:
        solution = read_solution("./example/solutions/"+tf+".log")
    score_output=""
    for i in range(len(data)):
        for j in range(len(data[0])):
            for k in range(len(data[0,0])):
                war_ori = war = round(i*0.05,2)
                wma_ori = wma = round(j*0.05,2)
                wml_ori = wml = round(k*0.05,2)
                if(sommeA1):
                    if(i+j+k!=20):
                        continue
                    if(use_score):
                        matrix = np.array([wml, wma, war])
                        #print("matrix=",matrix)
                        r = np.dot(J,matrix)
                        #print("r=",r)
                        wma,war = r[1],r[2]
                        wml = search_score(debug_folder+"/debugParser"+tf+"_war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori)+".log", solution)
                        #print("Score :",wml," for case _war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori))
                        score_output += "Score="+str(wml)+" Rank="+str(int(data[i,j,k]))+"   for case _war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori)+"\n"
                    #print("war=",war," wma=",wma," wml=",wml)
                if(war_ori>=war_from and war_ori<=war_to and wma_ori>=wma_from and wma_ori<=wma_to and wml_ori>=wml_from and wml_ori<=wml_to):
                    v = data[i,j,k]
                    if(only_red):
                        if(int(v)==1):
                            couleur = 'r'
                            count+=1
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                    elif(only_black):
                        if(int(v)==20000):
                            couleur='black'
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                    elif(only_yellow):
                        if(int(v)==100000):
                            couleur='yellow'
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                    elif(int(v)>0 and int(v)<=100000):
                        if(int(v)==1):
                            couleur = 'r'
                            count+=1
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                        elif(int(v)==100000):
                            print("100000 ->",file)
                            couleur = 'yellow'
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                        elif(int(v)==20000):
                            if(not without_black):
                                couleur = 'black'
                                sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                        else:
                            couleur = int(v)
                            war_l.append(war)
                            wma_l.append(wma)
                            wml_l.append(wml)
                            #print("ccc ",couleur, v)
                            couleur_l.append(couleur)

                    """
                    if(tab[i,j,k]==1):
                        war = round(i*0.05,2)
                        wma = round(j*0.05,2)
                        wml = round(k*0.05,2)
                        ax1.scatter3D(war,wma,wml, c='g',marker='.')
                    elif(tab[i,j,k]==-1):
                        war = round(i*0.05,2)
                        wma = round(j*0.05,2)
                        wml = round(k*0.05,2)
                        ax1.scatter3D(war,wma,wml, c='r',marker='.')
                    """
    war_a = np.array(war_l)
    wma_a = np.array(wma_l)
    wml_a = np.array(wml_l)
    couleur_a = np.array(couleur_l)
    cmap = sns.light_palette('black',reverse=True,n_colors=1)
    cmap2 = sns.light_palette('blue',reverse=False,n_colors=999)
    cmap3 = sns.light_palette('red',reverse=False,n_colors=1)
    #cmap.extend(cmap2.extend(cmap3))
    #norm = mpl.colors.BoundaryNorm([-999, 0, 998, 999], cmp.N)
    sc = ax1.scatter3D(war_a,wma_a,wml_a, c=couleur_a, marker='.', cmap='winter')
    ax1.set_xlabel('war')
    ax1.set_ylabel('wma')
    ax1.set_zlabel('wml')
    if(file!="#"):
        plt.savefig(file)
    #plt.colorbar(sc)
    plt.show()
    with open('./files_scores/'+tf+".log", 'w+') as f_o:
        f_o.write(score_output)
    

            
            
def combiner(current_result, tab):
    for i in range(len(tab)):
        for j in range(len(tab[0])):
            for k in range(len(tab[0,0])):
                current_result[i,j,k]=min(tab[i,j,k],current_result[i,j,k])
    return current_result
```

```python
def read_solution(file):
    ret = []
    with open(file) as f:
        while True:
            lines = f.readline()
            if not lines:
                break
            ret.append(lines[0:-1])
    return ret


def check_solution_correct(file,solution):
    num=0
    last_score = "***"
    with open(file) as f:
        while True:
            lines = f.readline()
            if not lines:
                break
            if (lines[0]!="["):
                continue
            num += 1
            line = lines.split(" ")
            best = []
            for l in line[8:]:
                if("execution_end_time" in l or "mission_end_time" in l or "EndRoot" in l):
                    break
                best.append(l)
            if(line[1]==last_score):
                num -= 1
            last_score = line[1]
            #print("solution ",solution,"\n")
            #print("best ",best,"\n")
            if (operator.eq(best,solution)):
                #print("Really!")
                return num;
            
        return 2000;
    
def search_score_best(debug_file, solution):
    premier = 0
    try:
        with open(debug_file) as f:
            while True:
                lines = f.readline()
                if not lines:
                    break
                if (lines[0]!="["):
                    continue
                line = lines.split(" ")
                if(line[0]=="[1er]"):
                    premier = float(line[1])
                best = []
                for l in line[8:]:
                    if("execution_end_time" in l or "mission_end_time" in l or "EndRoot" in l):
                        break
                    best.append(l)
                #print("Trying : Best=",best, " Solution=",solution)
                if (operator.eq(best,solution)):
                    return float(line[1]), premier
            return 0, premier
    
    except FileNotFoundError:
        print("File not found! -> ",debug_file)
        
def search_diff_between_best_second(file):
    num=0
    last_score = "***"
    with open(file) as f:
        while True:
            lines = f.readline()
            if not lines:
                break
            if (lines[0]!="["):
                continue
            num += 1
            line = lines.split(" ")
            if(line[1]==last_score):
                num -= 1
            if (num==2):
                if(float(line[1])==0):
                    return 0
                return float(last_score)-float(line[1])
            last_score = line[1]
        return 0;
        
def dessin3D_best(data, file="#", dynamic=False, war_from=0, war_to=1, wma_from=0, wma_to=1, wml_from=0, wml_to=1, \
             only_red=False, only_black=False, without_black=False,\
             only_yellow=False, sommeA1=False, use_score=True, debug_folder="./files_debug", tf=""):
    if(dynamic):
        %matplotlib notebook
    else:
        %matplotlib inline
    J = np.array([[1,1,1],[0,1,0.5],[0,0,(3**0.5)/2]])
        
    fig = plt.figure()
    ax1 = plt.axes(projection='3d')
    war_l = []
    wma_l = []
    wml_l = []
    couleur_l = []
    count = 0
    if use_score:
        solution = read_solution("./example/solutions/"+tf+".log")
    score_output=""
    diff_premier_deuxieme = 0
    war_d,wma_d,wml_d = 0,0,0
    for i in range(len(data)):
        for j in range(len(data[0])):
            for k in range(len(data[0,0])):
                war_ori = war = round(i*0.05,2)
                wma_ori = wma = round(j*0.05,2)
                wml_ori = wml = round(k*0.05,2)
                if(sommeA1):
                    if(i+j+k!=20):
                        continue
                    if(use_score):
                        matrix = np.array([wml, wma, war])
                        #print("matrix=",matrix)
                        r = np.dot(J,matrix)
                        #print("r=",r)
                        wma,war = r[1],r[2]
                        wml, premier = search_score_best(debug_folder+"/debugParser"+tf+"_war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori)+".log", solution)
                        #print("Score :",wml," for case _war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori))
                        score_output += "Score="+str(wml)+" Rank="+str(int(data[i,j,k]))+" Premier_score="+str(premier)+"   for case _war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori)+"\n"
                    #print("war=",war," wma=",wma," wml=",wml)
                if(war_ori>=war_from and war_ori<=war_to and wma_ori>=wma_from and wma_ori<=wma_to and wml_ori>=wml_from and wml_ori<=wml_to):
                    v = data[i,j,k]
                    temp = search_diff_between_best_second(debug_folder+"/debugParser"+tf+"_war"+str(war_ori)+"_wma"+str(wma_ori)+"_wml"+str(wml_ori)+".log")
                    if temp>diff_premier_deuxieme:
                        diff_premier_deuxieme = temp
                        war_d,wma_d,wml_d = war_ori,wma_ori,wml_ori
                    #Afficher le meilleur score en un croix orange
                    #sc = ax1.scatter3D(war,wma,premier, c="orange", marker='x')
                    
                    if(only_red):
                        if(int(v)==1):
                            #chercher le score au rang 2 et mettre Ã  jour le diff_premier_dexieme
                            
                            couleur = 'r'
                            count+=1
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                    elif(only_black):
                        if(int(v)==20000):
                            couleur='black'
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                    elif(only_yellow):
                        if(int(v)==100000):
                            couleur='yellow'
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                    elif(int(v)>0 and int(v)<=100000):
                        if(int(v)==1):
                            couleur = 'r'
                            count+=1
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                        elif(int(v)==100000):
                            print("100000 ->",file)
                            couleur = 'yellow'
                            sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                        elif(int(v)==20000):
                            if(not without_black):
                                couleur = 'black'
                                sc = ax1.scatter3D(war,wma,wml, c=couleur, marker='.')
                        else:
                            couleur = int(v)
                            war_l.append(war)
                            wma_l.append(wma)
                            wml_l.append(wml)
                            #print("ccc ",couleur, v)
                            couleur_l.append(couleur)

                    """
                    if(tab[i,j,k]==1):
                        war = round(i*0.05,2)
                        wma = round(j*0.05,2)
                        wml = round(k*0.05,2)
                        ax1.scatter3D(war,wma,wml, c='g',marker='.')
                    elif(tab[i,j,k]==-1):
                        war = round(i*0.05,2)
                        wma = round(j*0.05,2)
                        wml = round(k*0.05,2)
                        ax1.scatter3D(war,wma,wml, c='r',marker='.')
                    """
    war_a = np.array(war_l)
    wma_a = np.array(wma_l)
    wml_a = np.array(wml_l)
    couleur_a = np.array(couleur_l)
    cmap = sns.light_palette('black',reverse=True,n_colors=1)
    cmap2 = sns.light_palette('blue',reverse=False,n_colors=999)
    cmap3 = sns.light_palette('red',reverse=False,n_colors=1)
    #cmap.extend(cmap2.extend(cmap3))
    #norm = mpl.colors.BoundaryNorm([-999, 0, 998, 999], cmp.N)
    sc = ax1.scatter3D(war_a,wma_a,wml_a, c=couleur_a, marker='.', cmap='winter')
    ax1.set_xlabel('war')
    ax1.set_ylabel('wma')
    ax1.set_zlabel('wml')
    if(file!="#"):
        plt.savefig(file)
    #plt.colorbar(sc)
    plt.show()
    score_output += "\n\nIn this case, the biggest difference between the first solution and the second is "+ str(diff_premier_deuxieme)+" when _war"+str(war_d)+"_wma"+str(wma_d)+"_wml"+str(wml_d)+"\n"
    with open('./files_scores/'+tf+".log", 'w+') as f_o:
        f_o.write(score_output)
    

            
            
def combiner(current_result, tab):
    for i in range(len(tab)):
        for j in range(len(tab[0])):
            for k in range(len(tab[0,0])):
                current_result[i,j,k]=min(tab[i,j,k],current_result[i,j,k])
    return current_result
```
